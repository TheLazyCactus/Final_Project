```python
#Loop to get all the page
import time
response_1 = []
for x in range(1,14):
    params = {
    'noRedirect': '0',
    'page': str(x) ,
    }
    response = requests.get(
    'https://www.carrefour.fr/r/cremerie/beurres-cremes',
    params=params,
    cookies=cookies,
    headers=headers,
    )
    response_1.append(response.json())
    time.sleep(2)

print(response_1)
```

```python
#Get the relevant columns and extract the price/kg
import pandas as pd

df_1 = pd.DataFrame()  # Initialize an empty DataFrame

for item in response_1:  # Loop through each JSON object
    try:
        # Ensure 'data' exists and extract it safely
        df_data = pd.DataFrame(item.get('data', []))  
        
        # Ensure 'attributes' exists before normalizing
        if 'attributes' in df_data:
            df_product = pd.json_normalize(df_data['attributes'])
        else:
            print("Warning: 'attributes' key missing in data.")
            continue  # Skip this item if 'attributes' is missing

        price_unit = []
        for x in df_product.get("ean", []):  # Use .get() to avoid KeyError
            try:
                ean_index = df_product[df_product["ean"] == x].index[0]
                price_column = f"offers.{x}.0261-150-6.attributes.price.perUnitLabel"

                # Check if the column exists before accessing
                if price_column in df_product.columns:
                    price_unit_variable = df_product.loc[ean_index, price_column]
                else:
                    price_unit_variable = None  # Default if price column missing

                price_unit.append(price_unit_variable)
            except Exception as e:
                print(f"Error processing EAN {x}: {e}")
                price_unit.append(None)  # Append None for failed entries

        price = []
        for x in df_product.get("ean", []):  # Use .get() to avoid KeyError
            try:
                ean_index = df_product[df_product["ean"] == x].index[0]
                price_column = f"offers.{x}.0261-150-6.attributes.price.price"

                # Check if the column exists before accessing
                if price_column in df_product.columns:
                    price_variable = df_product.loc[ean_index, price_column]
                else:
                    price_variable = None  # Default if price column missing

                price.append(price_variable)
            except Exception as e:
                print(f"Error processing EAN {x}: {e}")
                price.append(None)  # Append None for failed entries

        # Select relevant columns safely
        try:
            df_final = df_product[["title", "slug", "topCategoryName", "origin"]].copy()
        except KeyError as e:
            print(f"Missing column error: {e}")
            continue  # Skip this item if required columns are missing

        # Add price_unit column
        df_final["price_unit"] = price_unit
        df_final["price"] = price

        # Concatenate safely
        df_1 = pd.concat([df_1, df_final], ignore_index=True)

    except Exception as e:
        print(f"Error processing item: {e}")
        continue  # Skip the item on error

# Print the final DataFrame
print(df_1)
```
