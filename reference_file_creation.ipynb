```python
#check duplicate
import json
from collections import Counter
def extract_names(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        data = json.load(file)
        # Access 'data' key, which contains a list of dictionaries
        return [item['name'] for item in data['data']]  # Extract 'name' from each item in 'data'

# List of JSON file paths
json_files = [r"D:\Documents\GitHub\Final_Project\food.json", r"D:\Documents\GitHub\Final_Project\boisson.json", r"D:\Documents\GitHub\Final_Project\veggies.json"]

# Initialize an empty list to hold all names
all_names = []

# Extract names from each file and add to all_names
for file in json_files:
    all_names.extend(extract_names(file))

# Count occurrences of each name using Counter
name_counts = Counter(all_names)

unique_names = list(set(all_names))

print("Unique names:", unique_names)

# Duplicates value
duplicates = [name for name, count in name_counts.items() if count > 1]

print("Duplicate names:", duplicates)
```

```python
#merge 3 json in one
# List of JSON file paths
json_files = [r"D:\Documents\GitHub\Final_Project\food.json", r"D:\Documents\GitHub\Final_Project\boisson_unique.json", r"D:\Documents\GitHub\Final_Project\veggies_unique.json"]

# Initialize an empty list to store merged data
merged_data = []

# Read each file and extract 'data' key contents
for file in json_files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
        merged_data.extend(data['data'])  # Append the list inside 'data' key

# Save the merged data into a new JSON file
merged_json = {"data": merged_data}  # Wrap in 'data' key

with open("merged.json", "w", encoding="utf-8") as f:
    json.dump(merged_json, f, indent=4, ensure_ascii=False)  # Pretty print with UTF-8 support

print("Merged JSON saved to merged.json")
```

```python
#Transform JSON to CSV
import json
import csv

# Load JSON file
with open(r"D:\Documents\GitHub\Final_Project\merged.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Extract list of dictionaries from "data" key
items = data["data"]

csv_file = "reference.csv"

headers = ["name", "slug", "ecv", "footprint", "footprintDetail_id", "footprintDetail_value"]

with open(csv_file, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(headers)  # Write column names

    # Iterate over each item in the JSON
    for item in items:
        # If 'footprintDetail' exists and is a list, create multiple rows
        if "footprintDetail" in item and isinstance(item["footprintDetail"], list):
            for detail in item["footprintDetail"]:
                writer.writerow([
                    item["name"],
                    item["slug"],
                    item["ecv"],
                    item["footprint"],
                    detail["id"],
                    detail["value"]
                ])
        else:
            # Write a single row if 'footprintDetail' is missing or empty
            writer.writerow([
                item["name"],
                item["slug"],
                item["ecv"],
                item["footprint"],
                None,
                None
            ])

print(f"CSV file '{csv_file}' created successfully!")
```

```python
df["slug"] = df["slug"].replace({"champignonmorille": "champignon"})
```

```python
df_pivot = df.pivot_table(index=['slug', 'footprint'],
                          columns='footprintdetail_id',
                          values='footprintdetail_value',
                          aggfunc='first')  # or use another aggregation if needed

# Reset the index to bring 'slug' and 'footprint' as columns
df_pivot.reset_index(inplace=True)

# Rename columns for clarity
df_pivot.columns = [f'footprintdetail_value_{col}' if isinstance(col, int) else col for col in df_pivot.columns]

# Group by 'slug' and aggregate, if necessary
df_grouped = df_pivot.groupby('slug', as_index=False).agg(lambda x: x.iloc[0])  # Aggregating using the first value

print(df_grouped)
```
